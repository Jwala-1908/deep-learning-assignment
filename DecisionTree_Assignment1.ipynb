```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
train_df = pd.read_csv('D:/BITS/DL/Assignments/Assignment1/mnist_train.csv')
test_df = pd.read_csv('D:/BITS/DL/Assignments/Assignment1/mnist_test.csv')
```

```python
x_train,y_train = train_df.iloc[:,1:].values, train_df.iloc[:,0].values
x_test,y_test = test_df.iloc[:,1:].values, test_df.iloc[:,0].values
```

```python
clf = DecisionTreeClassifier(criterion='entropy',min_samples_leaf=2,min_weight_fraction_leaf=0.0001,max_depth=15,min_samples_split=4)
```

```python
clf.fit(x_train,y_train)
```

```python
prediction_test = clf.predict(x_test)
```

```python
acc = accuracy_score(y_test,prediction_test)
acc
```

```python
importances = clf.feature_importances_
ft = clf.feature_importances_
ft = ft*25500
plt.imshow(ft.reshape((28,28)))
plt.title("Feature Importances learnt via Decision Tree\ncriterion: Entropy,\nTest Accuracy={}%".format(acc*100))
```

```python
sns.heatmap(confusion_matrix(y_test,prediction_test),cmap='Blues_r',fmt='',annot=True,cbar=False)
plt.ylabel("Correct Label")
plt.xlabel("Predicted Label")
plt.title("Confusion Matrix for Decision Tree\nModel Accuracy={}%".format(acc*100))
```

```python
print("Classification Report for Decision Tree:\n{}".format(classification_report(y_test,prediction_test)))
```

```python
indices = np.argsort(importances)[::-1]
```

```python
for f in range(x_train.shape[1]):
    print("{} Feature {} : {}".format(f+1,indices[f],importances[indices[f]]))
```
